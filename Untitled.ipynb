```python
fig = plt.figure()
ax = plt.axes(projection="3d")

def z_function(x, y):
    
    z= (5*np.exp(-np.square(x-4)/(2*(1)**2)))*(3*np.exp(-np.square(y-3)/(2*(1)**2)))
    
    return z


z= np.zeros((len(x), len(y)))

x = np.linspace(0, 8, 20)
y = np.linspace(0, 20, 20)

print(x)
print(y)


X, Y = np.meshgrid(x, y)
Z = z_function(X,Y)

print(Z)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('z')

plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
ax.set_title('surface');
```
